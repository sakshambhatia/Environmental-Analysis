```python
import numpy as np 
import pandas as pd 
```

```python
data=pd.read_csv('/Users/sakshambhatia/Downloads/data.csv',encoding = "ISO-8859-1")
data.head()
```

```python
data.info()
```

```python
data.columns
```

```python
useful_column = {'sampling_date', 'state', 'location', 'type',
       'so2', 'no2', 'rspm', 'spm', 'pm2_5', 'date'}
```

```python
data = data[useful_column]
```

```python
replacements = {
   'state': {
      r'Uttaranchal': 'Uttarakhand', 
   }
}

data.replace(replacements, regex=True, inplace=True)
```

```python
data.isnull().sum()
```

```python
data.drop(columns=['pm2_5'],inplace=True)
```

```python
data['type'].fillna('Unknown', inplace=True)
```

```python
data_state = data[['state', 'so2', 'no2', 'rspm', 'spm']].groupby(['state'])
states = data_state.groups.keys()
mean_data_state = data_state.mean()
mean_data_state.reset_index(level=0,inplace=True)
mean_data_state
```

```python
data.drop(data[data['state']=='andaman-and-nicobar-islands'].index,inplace=True)
data.drop(data[data['state']=='Tripura'].index,inplace=True)
data.drop(data[data['state']=='Lakshadweep'].index,inplace=True)
```

```python
gasses = {'so2', 'no2', 'rspm', 'spm'}
for gas in gasses:
    data[gas].fillna(data[gas].mean(),inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.to_csv("clean_data.csv")
```
